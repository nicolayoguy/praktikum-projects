- E - Everyone

- Вторая гипотеза – что это игры не из американского региона.

Проверим обе.

```python
# построим долю игр без рейтинга по годам
fig, ax = plt.subplots()

df.groupby('release_year')['rating'].agg(lambda x: x.isna().sum()/len(x)).plot(ax=ax)

ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.set_xlabel('Годы')
ax.set_ylabel('Доля игр без рейтинга')
ax.set_title('Доля игр без рейтинга по годам')

plt.show()
```

```python
# определим долю игр без рейтинга для игр, предназначенных для разных регионов
# будем считать, что игра локальная – если у неё доля продаж 
# в Америке, Европе, Японии или других регионах больше 75% от суммарных
# остальные игры запишем как global. добавим эти значения в колонку game_region

def set_game_region(row):
    if not row.total_sales:
        return np.nan
    sales = [row.na_sales, row.eu_sales, row.jp_sales, row.other_sales]
    names = ['na', 'eu', 'jp', 'other']
    for region, name in zip(sales, names):
        if region:
            if (region / row.total_sales) >= 0.75:
                return name
    return 'global'

df['game_region'] = df.apply(set_game_region, axis=1)
game_region_share = df.game_region.value_counts().apply(lambda x: x / len(df))


(df
 .groupby('game_region')['rating']
 .agg([lambda x: x.isna().sum()/len(x),
      lambda x: x.isna().sum()])
 .join(game_region_share)
 .rename({'<lambda_1>': 'no_rating_count',
         '<lambda_0>': 'no_rating_share',
         'game_region': 'games_share'},
         axis=1)
 .style
 .format({'no_rating_share': '{:.0%}',
         'games_share': '{:.0%}'})
 .background_gradient(cmap='Reds',
                      axis=0,
                     subset=['no_rating_share', 'games_share'])
)
```

```python
# и совместим обе гипотезы
fig, ax = plt.subplots()

(df
 .pivot_table(index='game_region',
              columns='release_year',
              values='rating',
              aggfunc=lambda x: x.isna().sum()/len(x))
 .T
 .plot(ax=ax)
)

# ax.xaxis.set_major_locator(MultipleLocator(10))
# ax.yaxis.set_major_locator(MultipleLocator(5))
# ax.yaxis.set_minor_locator(AutoMinorLocator(50))
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.set_xlabel('Годы')
ax.set_ylabel('Доля игр без рейтинга')
ax.set_title('Доля игр без рейтинга среди игр на разных рынках')

plt.show()
```

Ну что ж - обе наших гипотезы подтвердились.

- Среди игр до 2000 года выпуска почти нет игр с рейтингом.
- У игр, предназаченных для Японии, почти нет рейтинга. А в других больших регионах доля безрейтинговых не превышает в среднем 20-30%.

```python
# для каждой игры посмотрим, какие оценки у неё есть, и нарисуем общий график
def set_score_type(row):
    no_user = (row.user_score != row.user_score)
    no_critic = (row.critic_score != row.critic_score)
    
    if no_user and no_critic:
        return 'no_score'
    elif no_user and not no_critic:
        return 'critic_only'
    elif not no_user and not no_critic:
        return 'critic_and_user'
    elif not no_user and no_critic:
        return 'user_only'

fig, ax = plt.subplots()
(df
 .apply(set_score_type, axis=1)
 .value_counts()
 .sort_index()
 .plot(kind='pie', ax=ax, colors=['g', '#20e80e', 'r', '#20e80e']))
ax.set_title('Распределение игр по наличию оценок критиков и пользователей')
ax.set_ylabel('')
plt.show()
```

Чтобы определять актуальность платформы не по графикам, а численно - рассчитаем процентный коэффициент жизненного цикла платформы. Поделим количество игр, вышедших на платформе в последний год - у нас это 2016, на максимальное значение выхода игр на этой платформе. У растущих платформ, последний год и будет максимумом – то есть коэффициент будет 100%. У платформ, прошедших свой пик, значение будет меньше 100% или вовсе 0%, если в 2016 году для платформы не выходило игр.

```python
# табличка с коэффициентами по платформам
life_stage_coef = (df
 .pivot_table(index='platform_full_name',
              columns='release_year',
              values='name',
              aggfunc='count')
.apply(lambda x: x.fillna(0).iloc[-1] / x.max(), axis=1)
)
life_stage_coef.name = 'stage_coef'
```

```python
# таблица с основными показателями по платформам
table = (df[df.company.isin(BIG3)]
 .groupby(['platform_type', 'company', 'platform_full_name'])[['release_year', 'name', 'total_sales']]
 .agg({'release_year': [lambda x: x.max() - x.min() + 1, 'max'],
      'name': 'count',
      'total_sales': ['sum', 'median']})
 .join(life_stage_coef, how='left')
 .rename({('release_year', '<lambda_0>'): 'life span',
          ('release_year', 'max'): 'last year',
          ('name', 'count'): 'total games',
          ('total_sales', 'sum'): 'total sales',
          ('total_sales', 'median'): 'median sale'}, axis=1)
 .sort_values(by=['platform_type', 'company', 'stage_coef', 'last year'], ascending=False)
 
)

# отображаем таблицу с форматированием
(table.style
 .format({'last year': "{:.0f}",
          'life span': "{:.0f}",
          'total sales': "{:.1f}",
          'median sale': "{:.2f}",
         'stage_coef': "{:.0%}",})
 .set_sticky(axis="index")
 .background_gradient(cmap='Reds',
                      axis=0,
                      subset=['life span', 'total sales', 'median sale', 'total games', 'stage_coef']))
```

```python
# посчитаем средние показатели для платформ, которые уже закончили свой жизненный цикл
(table
 .query('stage_coef == 0')
 .pivot_table(index='platform_type',
              values=['life span', 'total games', 'total sales', 'median sale']))
```

Опишем, что мы видим:
- Есть шесть платформ с коэффициентом цикла больше 30% – это и есть актуальное поколение.
- Среди уже неактуальных платформ среднее время жизни поколения составляет 10-11 лет (это срок, в течение которого выходили новые игры, а сама платформа, очевидно, живёт ещё несколько лет после окончания выхода новых игр). То есть можно предположить, что актуальные платформы сейчас находятся примерно в середине жизни – до заката осталось 4-6 лет.

Здесь и далее для рассмотрения продаж будем использовать среднее (а не медиану). Большинство наших срезов устроено так, что у нас есть большие выбросы вверх. Они не влияют на медиану, а среднее сдвигают. Но эти выбросы – то есть супер-хитовые игры с многомиллионными продажами - нас как раз и интересуют. Мы хотим уметь определять факторы, влияющие именно на их появление, так что логично использовать среднее.

```python
# для каждой игры посчитаем, сколько у неё платформ, компаний и типов платформ
# добавим эти столбцы к нашим данным
# и сгруппируем по количеству компаний, типов и платформ
games_platforms = df.groupby('name')[['platform', 'company', 'platform_type']].transform('nunique')
df = df.join(games_platforms, rsuffix='_count')

# и обновим наш срез
df_sales = df.query('release_year in @COMPARABLE_PERIOD and platform_full_name in @MODERN_PLATFORMS')
```

```python
(df
 .groupby(['company_count', 'platform_type_count', 'platform_count'])
 [['total_sales', 'name']]
 .agg({'total_sales': 'mean', 'name': ['count', 'nunique']})
 .style
 .format({('total_sales', 'mean'): "{:.2f}"})
 .background_gradient(cmap='Reds',axis=0)
)
```

Видим, что наше предположение для всех данных скорее оправдывается. Средние продажи на одной платформе у игр, которые выходили на нескольких платформах, выше, чем у одноплатформенных. 

Но разбивка по количеству компаний/типов/платформ слишком сложная. Для простоты сведём её к 4 случаям:
- только одна платформа
- несколько платформ одного типа одной компании 
- несколько платформ разного типа одной компании
- платформы разных компаний

```python
# добавим к данным столбец с типом мультиплатформенности
def set_multi_type(row):
    if row.platform_count == 1:
        return 'single'
    elif row.company_count == 1:
        if row.platform_type_count == 1:
            return 'one_company_multi_generation'
        else:
            return 'one_company_multi_platform'
    else:
        return 'multi_company'
    
df['multi_platform_type'] = df.apply(set_multi_type, axis=1)

# и добавим эти данные в наш срез
df_sales = df.query('release_year in @COMPARABLE_PERIOD and platform_full_name in @MODERN_PLATFORMS')
```

```python
# посчитаем средние продажи одной игры разных типов только для нашего среза актуальных платформ
(df_sales
 .groupby('multi_platform_type')
 [['total_sales', 'name']]
 .agg({'total_sales': 'mean', 'name': ['count', 'nunique']})
 .style
 .format({('total_sales', 'mean'): "{:.2f}"})
 .background_gradient(cmap='Reds', axis=0)
)
```

```python
# и проверим статистическую значимость отличия средних продаж игр,
# выпускавшихся на платформах разных компаний, от остальных
multi = df_sales.query('multi_platform_type == "multi_company"').total_sales.dropna()
others = df_sales.query('multi_platform_type != "multi_company"').total_sales.dropna()

ALPHA = 0.05
pvalue = st.ttest_ind(multi, others).pvalue
print(f'Различие средних значений статистически {"значимо" if pvalue < ALPHA else "не значимо"}. Pvalue: {pvalue:.6f}')
```

Ну что ж - наши данные говорят, что средние продажи игры, которая выходила на платформах разных компаний, выше, чем продажи других игр. 0.60 млн копий в среднем – для мульти-компаний и 0.46 млн для всех остальных. (Речь идёт о средних продажах на одной платформе, а не о совокупных продажах игры на нескольких платформах.) Это отличие статистически значимо – уровень pvalue - меньше 0.01.

```python
# распределение возрастных рейтингов среди игр разного жанра
genre_rating = (df_sales
                .pivot_table(index='genre',
                             columns=df_sales.rating.apply(lambda x: x if x else 'No rating'),
                             values=['total_sales'],
                             aggfunc=lambda x: x.sum())
                .apply(lambda col: col / col.sum(), axis=1)
               )

(genre_rating[[('total_sales', 'E'),
              ('total_sales', 'E10+'),
              ('total_sales', 'T'),
             ('total_sales', 'M'),
             ('total_sales', 'No rating')]]
 .style
 .format("{:.1%}")
 .background_gradient(cmap='Reds', axis=1)
)
```

Учитывая вышесказанное вряд ли стоит искать влияние рейтинга на продажи. Разве что можно посмотреть, различаются ли продажи игр одного жанра, но с разными возрастными ограничениями. Для примера возьмём жанры, где сопоставимые доли игр разных рейтингов. Посмотрим шутеры, ролевые игры и экшн. Узнаем, что лучше продаётся 13+ или 17+.

```python
# сравним продажи медианной игры жанра с рейтингом М и Т
genre = ['Shooter', 'Role-Playing', 'Action']
rating = ['T', 'M']
(df_sales.query('genre in @genre and rating in @rating')
 .pivot_table(index='genre',
              columns='rating',
              values='total_sales',
              aggfunc=['count', 'mean'])
 .style
 .format("{:.2f}")
 .background_gradient(cmap='Reds', axis=1, subset=['mean'])
)
```

```python
# проверим статистическую значимость различий
for gen in genre:
    genre_t = df_sales.query('genre == @gen and rating == "T"').total_sales
    genre_m = df_sales.query('genre == @gen and rating == "M"').total_sales
    pvalue = st.ttest_ind(genre_t, genre_m).pvalue
    print(f'{gen}: разница {"не значима" if pvalue > 0.05 else "значима"}. Pvalue: {pvalue:.5f}')
```

Средни значения продаж у игр одного жанра, но разного возрастного рейтинга различаются – у экшн-игр и RPG продажи взрослого контента больше, чем продажи игр с рейтингом 13+, а у шутеров – наоборот. Однако тест на статистическую значимость с уровнем alpha 0.05 проходят только два жанра Action и Role-Playing. Здесь мы можем считать, что полученная разница скорее неслучайна. С жанром Shooter полученные различия с высокой вероятностью могут быть отнесены на счёт случайности. Последнее, вероятно, объясняется малым числом игр для проверки гипотезы – шутеров с подростковым рейтингом в нашей выборке всего 14 штук.

## Проверка гипотез

Нас попросили проверить два предположения:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

Сделаем это.

### Пользовательские рейтинги Xbox One и PC

У платформ разное время жизни: игры на Windows выходят больше 30 лет, на Xbox One – всего 4. Сравнивать отзывы можно на все игры или только за сопоставимый период. Нам не уточнили, какой именно результат хотя получить, так что сделаем оба варианта.

Нулевая гипотеза – средние значения оценок в совокупностях не различаются, а наблюдаемые различия выборочных средних это случайность.

Альтернативная - средние значения в совокупностях различны (больше или меньше).

Уровень alpha пусть будет 0.05 - вероятность того, что мы признаем различными совокупности, которые на самом деле не различаются.

```python
# годы жизни Xbox One
min_year = int(df[df.platform_full_name == 'Xbox One'].release_year.agg('min'))
max_year = int(df[df.platform_full_name == 'Xbox One'].release_year.agg('max'))

xbox_years = list(range(min_year, max_year+1))
xbox_years
```

```python
# посчитаем средние оценки за сопоставимый период
display(df
         .query('release_year in @xbox_years and platform_full_name in ["Xbox One", "Microsoft Windows"]')
         .pivot_table(index='platform_full_name', values='user_score')
        )

xbox = df.query('platform_full_name == "Xbox One"').user_score.dropna()
windows = df.query('release_year in @xbox_years and platform_full_name == "Microsoft Windows"').user_score.dropna()

ALPHA = 0.05
pvalue = st.ttest_ind(xbox, windows, equal_var=False).pvalue
print(f'Различие средних значений статистически {"значимо" if pvalue < ALPHA else "не значимо"}. Pvalue: {pvalue:.6f}')
```

```python
# посчитаем средние оценки за полный период
display(df
         .query('platform_full_name in ["Xbox One", "Microsoft Windows"]')
         .pivot_table(index='platform_full_name', values='user_score')
        )

xbox = df.query('platform_full_name == "Xbox One"').user_score.dropna()
windows = df.query('platform_full_name == "Microsoft Windows"').user_score.dropna()

ALPHA = 0.05
pvalue = st.ttest_ind(xbox, windows, equal_var=False).pvalue
print(f'Различие средних значений статистически {"значимо" if pvalue < ALPHA else "не значимо"}. Pvalue: {pvalue:.6f}')
```

Ну что ж – как посчитаем, так и будет )))

Если сравнивать игры, вышедшие в одни и те же годы, то статистически значимой разницы между оценками пользователей нет. 

Если же брать игры за всю историю Windows, то оценки пользователей у них в среднем выше, чем у игр на Xbox One. Это отличие статистически значимо.

### Пользовательские рейтинги жанров Action и Sports

Сначала посмотрим, как различаются пользовательские оценки всех игр жанров Action и Sports.

```python
display(df
         .query('genre in ["Action", "Sports"]')
         .pivot_table(index='genre', values='user_score')
        )

action = df.query('genre == "Action"').user_score.dropna()
sports = df.query('genre == "Sports"').user_score.dropna()

ALPHA = 0.05
pvalue = st.ttest_ind(action, sports, equal_var=False).pvalue
print(f'Различие средних значений статистически {"значимо" if pvalue < ALPHA else "не значимо"}. Pvalue: {pvalue:.6f}')
```

Отвергнуь нулевую гипотезу, что средние значения совокупностей равны, а полученная разница выборочных средних это случайность, мы не можем. Различие средних значений пользовательских оценок игр двух жанров статистически не значимо.

Однако, кажется, что сравнивать всё скопом не совсем правильно. Платформы сильно различаются по жанрам, количеству оценок, поведению пользователей. Так что, возможно, правильнее было бы сравнивать оценки игр в рамках одной и той же платформы. Посмотрим.

```python
# посчитаем для каждой платформы средние значения оценок и статтест
# оставим только те платформы, для которых в каждом жанре есть не меньше 30 игр с оценками

platform_sport_action = []

for platform in df.platform_full_name.unique():
    try:
        data = df[df.platform_full_name == platform]
        action = data.query('genre == "Action"').user_score.dropna()
        sports = data.query('genre == "Sports"').user_score.dropna()
        if len(action) == 0 or len(sports) == 0:
            continue
        pvalue = st.ttest_ind(action, sports, equal_var=False).pvalue
        verdict = ("значимо" if pvalue < ALPHA else "не значимо")
        platform_sport_action.append([platform,
                                      action.mean(),
                                      sports.mean(), 
                                      pvalue,
                                      verdict,
                                      min(len(action), len(sports))]
                                    )

    except:
        continue
        
platform_sport_action = pd.DataFrame(data=platform_sport_action,
                                     columns=['platform',
                                              'action_mean',
                                              'sports_mean',
                                              'pvalue',
                                              'significance',
                                              'games_with_rating_min_count']
                                    )

(platform_sport_action
 .set_index('platform')
 .query('games_with_rating_min_count >= 30')
 .sort_values(by='pvalue')
 .style
 .background_gradient(cmap='Reds',
                      axis=1,
                      subset=['action_mean', 'sports_mean'])
 .format({'action_mean': "{:.2f}",
          'sports_mean': "{:.2f}"})
)
```

Есть 10 платформ, на которых в жанрах Action и Sports есть не менее 30 игр с пользовательскими оценками. На шести из этих платформ  между жанрами есть статистически значимое отличие оценок. В пяти из шести случаев пользователи выше оценивали игры жанра Action. Исключение – Xbox, где игрокам больше нравились спортивные игры, чем Action.
