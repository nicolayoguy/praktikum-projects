```python
# добавим столбец is_bad – если  

def set_bad(row):
    if row.rooms_number > 5:
        return set(['rooms_number'])
    else:
        return set()
    
df['is_bad'] = df.apply(set_bad, axis=1)
```

```python
# и сразу напишем функцию для фильтра по нашему столбцу с пометками плохих значений
def filter_bad_param(cell, param=[], all_mode=False):
    '''
    функция для фильтрации по столбцу is_bad.
    param - список значений из столбца is_bad, которые мы не хотим видеть в результирующем датасете.
    если не установлен параметр all_mode=True, то отфильтровываются любые пересечения.
    если all_mode=True, то отфильтровываются только строки, где есть все значения, переданные в param.
    
    пример запуска
    df[df.is_bad.apply(filter_bad_param, args=[['kitchen_area', 'kitchen_area_ratio'], True])]
    '''
    param = set(param)
    if not all_mode:
        if not cell & param:
            return True
        else:
            return False
    else:
        if not param.issubset(cell):
            return True
        else:
            return False
```

```python
# и подробнее сравним площадь и цену у однушек, студий и квартир с открытой планировкой

fig, (ax1, ax2) = plt.subplots(1 ,2)
fig.set_figwidth(10)
fig.set_figheight(8)

for param, ax in zip(['total_area', 'last_price'], [ax1, ax2]):
    studios = df.query('is_studio == True')[param].to_list()
    one_room = df.query('rooms_number == 1')[param].to_list()
    open_plan = df.query('is_open_plan == True and rooms_number in [0,1]')[param].to_list()
    ax.boxplot([studios, one_room, open_plan], labels=['Студии', 'Однушки', 'Открытая'])

ax1.set_title('Общая площадь')
ax1.set_ylabel('Общая площадь, кв. м')
ax1.set_ylim(0,80)
ax1.grid()

ax2.set_title('Цена')
ax2.set_ylabel('Цена')
ax2.set_ylim(0, 10_000_000)
ax2.grid()

fig.suptitle('Распределение студий, однушек и квартир с открытой планировкой по площади и цене', y=0.96, fontsize=16)
plt.show()
```

2. Строки с 20-30 метрами похожи на то, что ошиблись разрядами и должно быть 2-3 метра. Поделим значения на 10. И пометим эти строки как подозрительные по этому параметру.

3. Что делать с остальными, не очень понятно. Потолок 10 метров вряд ли возможен. 4-5 метров наверно могут быть – особенно, если там площади квартиры по несколько сотен метров. 3 квартиля плюс 1,5 размаха даёт слишком маленькое ограничение – 3,08 метра. В центре Петербурга явно могут быть реальные квариры с более высокими потолками. Тогда, руководствуясь здравым смыслом, **выставим предел хороших значений – 6 метров**. Всё что выше, оставим как есть, но пометим как подозрительные по высоте потолков.

4. Есть три строки с высотой потолков меньше 2 метров. Выставим им 2 метра. 

```python
# функция для расчёта границ хороших значений
def compute_outliers(param):
    Q1 = df[param].quantile(0.25)
    Q3 = df[param].quantile(0.75)
    IQR = Q3 - Q1
    return [Q1 - IQR, Q3 + IQR]
```

```python
ceiling_height_limit = compute_outliers("ceiling_height")[1]
print(f'Традиционный подход 3-ий квартиль плюс 1,5 размаха даёт слишком маленький предел для нащих данных - {ceiling_height_limit:.2f} м')
```

```python
def fix_ceiling_height(row):
    height = row.ceiling_height
    param = 'ceiling_height'
    
    if height == 100:
        row.ceiling_height = None
        row.is_bad.add(param)
        return row
    elif 20 <= height <= 32:
        row.ceiling_height = round(height / 10, 2)
        row.is_bad.add(param)
        return row
    elif 6 < height < 20:
        row.is_bad.add(param)
        return row
    elif height < 2:
        row.ceiling_height = 2
        row.is_bad.add(param)
        return row
    
    return row
```

```python
df = df.apply(fix_ceiling_height, axis=1)
```

60 – это, вероятно, опечатка – должно быть 6. Квартира из объявления – на 4-м этаже.<br>52 – непонятно. Квартира на 18-м.

Посмотрим, нет ли объвлений из таких же домов. Кажется, что дом достаточно хорошо определяется парой значений *расстояние до центра* и *расстояние до аэропорта*. Поищем, нет ли объявлений с такими же расстояниями.

Итого:
- 60-этажный заменим на 6-этажный. И пометим плохим по этажности.
- 52 этажа – заменим на None. И пометим плохим по этажности.
- Добавим колонку с типом этажности: первый, последний, другой.
- Запомним, что дома из объявлений можно определять по расстояниям до центра и аэропорта. И в рамках одного дома можно проверять разброс этажности, высоты потолков, цены квадратного метра и т.п.

```python
def fix_floors(row):
    floor_number = row.floor_number
    floors_total = row.floors_total
    param = 'floors_total'
    
    if floors_total == 60:
        row.floors_total = 6
        row.is_bad.add(param)
    elif floors_total == 52:
        row.floors_total = np.nan
        row.is_bad.add(param)
    
    if floor_number == 1:
        row['floor_type'] = 'первый'
    elif floor_number == floors_total:
        row['floor_type'] = 'последний'
    elif (floors_total != floors_total) or (floor_number != floor_number):  # если у нас не первый этаж и нет данных о высоте дома,
                                        # то мы не можем достоверно определить тип этажа – ставим "нет данных"
        row['floor_type'] = np.nan
    else:
        row['floor_type'] = 'другой'
    
    return row
```

```python
# сделаем колонку floor_type категорией
floor_type = pd.CategoricalDtype(categories=["первый", "другой", "последний"], ordered=True)
df['floor_type'] = np.nan
df = df.apply(fix_floors, axis=1)
df['floor_type'] = df['floor_type'].astype(floor_type)
```

```python
# посмотрим на гистограммы и отметим на них части, которые мы считаем подозрительными

fig, (ax1, ax2) = plt.subplots(1 ,2)
fig.set_figwidth(12)
fig.set_figheight(8)

for param, ax in zip(['living_area_ratio', 'kitchen_area_ratio'], [ax1, ax2]):
    ax.hist(df[param], bins=20)
    low, high   = compute_outliers(param) 
    ax.vlines(low, 0, 2000, color='red', linestyle='--')
    ax.vlines(high, 0, 2000, color='red', linestyle='--')
    ax.text(low, 2000, 'выбросы', ha='right')
    ax.text(high, 2000, 'выбросы')
    ax.set_ylabel('Количество квартир')
    ax.set_ylim(0, 6000)
    ax.set_xlim(0, 1)

ax1.set_title('Доля жилой площади от общей')

ax2.set_title('Доля площади кухни от общей')
ax2.yaxis.set_label_position("right")
ax2.yaxis.tick_right()

plt.show()

print('Границы нормальных значений:')
for param in ['living_area_ratio', 'kitchen_area_ratio']:
    low, high = compute_outliers(param)

    print(f'\tдля {param}: {low:.0%} - {high:.0%}')
```

```python
# пометим плохие значения

outlier_lims = {param: compute_outliers(param) for param in ['living_area_ratio', 'kitchen_area_ratio']}

def set_bad_area_ratio(row):
    cell = row['is_bad']
    # проверяем, не выходит ли какой-либо из параметров за пределы хороших значений
    for param in ['living_area_ratio', 'kitchen_area_ratio']:
        if row[param] != row[param]:
            return cell
        elif not outlier_lims[param][0] < row[param] < outlier_lims[param][1]:
            cell.add(param)
    # дополнительно проверяем, не получается ли в сумме двух площадей больше единицы
    if row['living_area_ratio'] + row['kitchen_area_ratio'] > 1:
        cell.update(['living_area_ratio', 'kitchen_area_ratio'])
    return cell

df['is_bad'] = df.apply(set_bad_area_ratio, axis=1)
```

Нуу, весенний и осенний всплески активности похожи на правду. Пик в июне – это, видимо, эффекты от продвижения.

И раз уж написали громоздкий код, используем его, чтобы посчитать то же самое по дням недели.

```python
# в один датасет положим количество публикаций по годам-неделям, а в другой по годам-неделям-дням
weeks = (df.groupby(['ad_year',
                     df.first_day_exposition.dt.weekofyear])
         ['ad_id']
         .count()
         .rename_axis(['year', 'week'])
         .reset_index()
        )

days = (df.groupby(['ad_year', 
                    df.first_day_exposition.dt.weekofyear, 
                    'ad_week_day'])
          ['ad_id']
          .count()
          .rename_axis(['year', 'week', 'day'])
          .reset_index()
         )

# к годам-неделям-дням допишем суммы за год-неделю и посчитаем долю дня от соответствующей недели
weeks_days = days.merge(weeks,
                        how='left',
                        on=['year', 'week'],
                        suffixes=['_day', '_week'])

weeks_days['day_share'] = weeks_days['ad_id_day'] / weeks_days['ad_id_week']
```

```python
# ещё добавим столбец, в котором посчитаем количество дней, в которые были значения.
# по нему будем отфильтровывать полные недели
full_week = weeks_days.groupby(['year', 'week'])['day_share'].count()

weeks_days = weeks_days.merge(full_week,
                              how='left',
                              on=['year', 'week'],
                              suffixes=[None, '_count'])
```

```python
# построим график по данным за полные недели 
fig, ax = plt.subplots(figsize=(10, 6))

(weeks_days.query('day_share_count == 7')
 .groupby('day')['day_share']
 .mean()
 .plot(kind='bar', ax=ax)
)

ax.set_xlabel('')
ax.set_ylabel('Доля от всех объявлений за неделю')
ax.set_title('Публикация объявлений по дням недели')
ax.set_xticklabels(['пн', 'вт', 'ср', 'чт', 'пт', 'сб', 'вс'])

plt.show()
```

Динамика по дням недели похожа на реальную – выходные и будние дни чётко видны.

Районов города у нас в данных нет, но, опираясь на расстояние до центра и цену квадратного метра, мы можем разделить город на центральный регион и условную окраину. Будем считать, что граница между этими регионами  будем проходить там, где цена жилья резко возрастает.

```python
# создадим колонку с расстоянием в километрах.
# у нас будет округление вверх. 1 км – это всё от нуля до 1.
df['dist_to_center_km'] = df.dist_to_center // 1000 + 1
```

```python
# нарисуем медианную цену для каждого километра удаленности от центра
# резкий рост цен видим при переходе от 9 к 8 км. 
# все что меньше 9 и будет нашим центром. закрасим эти колонки другим цветом.

spb_by_dist = (df[(df.district == 'spb_center') &
                  (df.is_bad == set())]
               .groupby('dist_to_center_km')['sq_m_price']
               .median()
              )

fig, ax = plt.subplots(figsize=(10, 6))
colors = ['coral' if km < 9 else 'blue' for km in spb_by_dist.index]
ax.bar(spb_by_dist.index, spb_by_dist.values, color=colors)

ax.set_xlabel('Расстояние до центра Петербурга, км')
ax.set_ylabel('Медианная цена квадратного метра, рублей')
ax.set_title('Цена квадратного метра жилья в зависимости от удаленности от центра города')
ax.grid(ls='--')

plt.show()
```

```python
# у нас что-то много колонок с типом регионов, но сделаем ещё одну )))
# в ней будет наше деление: центр, окраина, пригород, область

location_group = pd.CategoricalDtype(categories=["центр СПБ", 
                                                 "окраина СПБ",
                                                 "пригороды СПБ",
                                                 "Ленобласть"], ordered=True)

def set_location_group(row):
    district = row.district
    km = row.dist_to_center_km
    region = row.region
    
    if region == 'Ленинградская область':
        return 'Ленобласть'
    elif region == 'Санкт-Петербург' and district != 'spb_center':
        return 'пригороды СПБ'
    elif district == 'spb_center' and km >= 9:
        return 'окраина СПБ'
    elif district == 'spb_center' and km < 9:
        return 'центр СПБ'
    else:
        return None

df['location_group'] = df.apply(set_location_group, axis=1).astype(location_group)
```

## Другие факторы, коррелирующие с ценой квадратного метра

Очевидно, что помимо географии существуют и другие факторы, влияющие или коррелирующие с ценой недвижимости. Факторы, которые изменяют цену жилья в рамках одного типа локациии. Попробуем их выделить.

- Всё считаем для четырёх географических срезов. Ожидаем, что в них будет разное поведение.
- Без строк, в которых есть сомнительные данные. Они могут исказить или зашумить наши корреляции.
- Без объявлений, которые ещё стоят на продаже. Предполагаем, что финальная цена может заметно отличаться от предыдущих.

```python
# сделаем сет с чистыми данными

# зададим параметры, по которым мы не хотим видеть подозрительные значения 
params = [
    'balcony_number',
    'ceiling_height',
    'floors_total',
    'kitchen_area_ratio',
    'living_area_ratio',
    'locality_name',
    'rooms_number',
    'total_images'
        ]

mask = df.is_bad.apply(filter_bad_param, args=[params])

# отфильтруем из датасета объявления, которые ещё не проданы
# и объявления, в которых есть строки с плохими параметрами
clean_set = df[mask].query('ad_status == "sold"')

# и ещё уберём три строки из Ленобласти, у которых единственных в регионе почему-то есть расстояния
# из-за них в корреляциях для Ленобласти появляются выбросы
clean_set = clean_set[~((clean_set.location_group == "Ленобласть") &
         (clean_set.dist_to_airport == clean_set.dist_to_airport))]

# посмотрим, сколько строк мы отфильтровали
before_after = (clean_set.groupby('location_group')['ad_id'].count().reset_index()
                .merge(df.groupby('location_group')['ad_id'].count().reset_index(), on='location_group')
                .rename({'ad_id_x': 'чистые данные', 'ad_id_y': 'все данные'}, axis='columns')
                )
before_after['изменение'] = before_after['чистые данные'] / before_after['все данные'] - 1
before_after.style.format({'изменение': '{:.0%}'})
```

### Количественные факторы

Для колонок с количественными параметрами просто посчитаем корреляцию с ценой квадратного метра. Ради интереса будем смотреть все колонки.

```python
# группу, по которой будем сортировать корреляции, вынесем в переменную
group_to_sort_by = 'центр СПБ'
# group_to_sort_by = 'окраина СПБ'
# group_to_sort_by = 'пригороды СПБ'
# group_to_sort_by = 'Ленобласть'


# сделаем датасет, в котором значения корреляции будут только положительными,
# а пропуски будут заменены на нули
# по этому сету будем раскрашивать нашу таблицу
color_set = np.array(clean_set
                     .select_dtypes(include=['int', 'float', 'bool', 'category'])
                     .groupby('location_group')
                     .agg(lambda column: abs(column.corr(clean_set['sq_m_price'])))
                     .fillna(0)
                     .T
                     .sort_values(by=group_to_sort_by, ascending=False)
                    )

# строим таблицу с корреляциями. 
# сортировка будет по абсолютному значению корреляции – сильная будет сверху.
(clean_set
 .select_dtypes(include=['int', 'float', 'bool', 'category'])
 .groupby('location_group')
 .agg(lambda column: column.corr(clean_set['sq_m_price']))
 .T
 .sort_values(by=group_to_sort_by, key=lambda col: abs(col), ascending=False)
 .style
 .format("{:.2f}")
 .background_gradient(cmap='Reds', axis=None, gmap=color_set)
)
```

Факторы, влияющие на цену в рамках одной и той же группы, различаются между разными группами. Например, в центре Петербурга и пригородах расстояние до центра не оказывает влияния на цену квадратного метра. А для окраины города это может быть важно. Рассмотрим коррелирующие факторы по группам.

В центре города сильнее всего с ценой квадратного метра коррелируют размер кухни и общая площадь объекта. Можно предположить, что эти параметры отличают хорошие квартиры в старом фонде – до начала 20 века постройки, от плохих – советского времени постройки.

Для объектов на условной окраине города, вероятно, также очень важен тип дома – здесь выбор стоит между новостройкой и советским фондом. Новостройки характеризуются большой этажностью дома и большими кухнями. В советском фонде, соотвественно, всё наоборот. 

Также для окраины важна близость к центру. Чем ближе – тем дороже.

Интересно, что в центре города почти нет "скидки за опт" – с увеличением количества комнат и площади цена квадратного метра мало снижается. На окраине есть обратная корреляция между количеством комнат и ценой метра – чем больше комнат, тем дешевле выходит метр.

В пригородах Петербурга единственный параметр, для которого можно говорить о корреялции с ценой квадратного метра – это размер кухни. Цена квадратного метра выше в квартирах, где больше площадь кухни относительно общей площади.

Ленобласть похожа на окраину Петербурга – здесь дороже всего квартиры в многоэтажных домах – видимо, это новостройки с относительно большими кухнями и высокими потолками.


### Неколичественные факторы

Посмотрим, как влияет на цену квадратного метра этаж квартиры (первый-последний) и как менялась цена продажи во времени. 

```python
# посмотрим, как меняются цены в зависимости от этажа

# вектор с медианными ценами по регионам
median_prices = (clean_set.pivot_table(columns='location_group',
                                values='sq_m_price',
                                aggfunc='median')
          .loc['sq_m_price'])

# посчитаем медианные цены для разных этажей и сравним с медианой по всем
(clean_set
 .pivot_table(index='floor_type',
              columns='location_group',
              values='sq_m_price',
              aggfunc='median')
 .apply((lambda row: (row / median_prices) - 1), axis=1)
 .loc[['последний', 'другой', 'первый'], :]
 .style
 .format("{:.2%}")
)
```

Квартиры на первых и последних этажах продаются с дисконтом к средней цене. 

В пределах Петербурга и в Ленобласти расположение на первом этаже это серьёзный минус – из-за него цена квадратного метра снижается больше чем на 10%. В пригородах Петербурга первый этаж не так страшен – примерно 5% снижения.

В Петербурге последний этаж снижает цену только на 5%. В пригородах скидки почти нет. А вот в Ленобласти последний этаж в смысле цены это почти так же плохо как и первый.

Посмотрим, как менялись цены на недвижимость в течение нашего периода данных.

Поскольку у нас есть данные только о финальной цене – при которой объявление было снято с публикации, то считаем целесообразным эту цену привязывать не к дате создания объявления, а к дате снятия. Разница есть, учитывая, что для некоторых квартир срок экспозиции исчисляется годами.

```python
# функция для расчёта года-месяца продажи
def set_sale_year_month(row):
    first = row['first_day_exposition']
    days = row['days_exposition']
    try:
        sale_date = (first + pd.Timedelta(days, unit='day')).to_period('M')
        return sale_date
    except:
        return np.nan

# добавим колонку в исходный сет. и перепишем чистый сет
df['sale_year_month'] = df.apply(set_sale_year_month, axis=1)
clean_set = df[mask].query('ad_status == "sold"')
```

```python
# сделаем сводную таблицу изменения цены продажи по месяцам

# именованные функции для расчёта квантилей
def quant_25(row):
    return np.quantile(row, 0.25)

def quant_75(row):
    return np.quantile(row, 0.75)

price_pivot = clean_set.pivot_table(index='sale_year_month',
                                    columns='location_group',
                                    values='sq_m_price',
                                    aggfunc=['median', 
                                             quant_25,
                                             quant_75]
                                    )
```

```python
price_pivot.index
```

```python
# сделаем график изменения медианной цены.
# областями также покажем границы 1 и 3 квантиля

fig, ax = plt.subplots(figsize=(12, 8))

for name in price_pivot.quant_25: 
    ax.plot_date(price_pivot.index, price_pivot['median', name], fmt='o-', label=name)
    ax.fill_between(price_pivot.index, price_pivot['quant_25', name],
                    price_pivot['quant_75', name],
                    alpha=0.3)

ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')
ax.legend(loc='lower left')
ax.set_xlabel('')
ax.set_ylim(0, 150_000)
ax.set_ylabel('Цена квадратного метра на момент продажи')
ax.set_title('Изменение стоимости продажи квадратного метра по регионам')

plt.show()
```

```python
# оценим, как изменилась цена продажи
# сравним усредненные данные за первые и последние полгода
price_pivot['median'].tail(6).mean() / price_pivot['median'].head(6).mean() - 1
```
